```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("tips.csv")
df.head()
```

```python
total_bill = list(df['total_bill'])
```

```python
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
mean,std
```

```python
normalised_data=[]
for i in total_bill:
    normalised_data.append((i-mean)/std)
print(normalised_data)
```

```python
sns.histplot(total_bill,kde=True)
```

```python
sns.histplot(normalised_data,kde=True)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
Scaler=StandardScaler()
```

```python
Scaler.fit(df[['total_bill','tip']])
```

```python
df1 = pd.DataFrame(Scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tips'])
```
